Let's structure Phase 4 properly before writing any code. Based on everything EDA revealed, here's what the deep-dive will cover:Four analyses, each building on what EDA told us:

# [Phase 4 Plan](assets/html/phase4_plan.html)

**A — RFM segmentation** answers the customer question EDA raised: we know the top 10% of customers generate 82.6% of profit, but we don't yet know how many of those customers are at risk of churning. RFM gives each customer a score based on how recently they bought, how often they buy, and how much they spend — then groups them into actionable segments like Champions, Loyal, At Risk, and Lost.

**B — Cohort analysis** adds the time dimension EDA couldn't show. We know the business grew 89% in profit from 2014 to 2017 — but is that driven by customers who return repeatedly, or by constant acquisition of new ones? Cohort analysis tracks what percentage of customers acquired in each quarter are still buying in later quarters, and how much value each acquisition wave generates over time.

**C — Margin erosion** translates the discount findings from Q3 and Q4 into hard dollars. We know discounting has an r = −0.864 correlation with margin. This analysis simulates the counterfactual: if every transaction above 20% discount had been capped at 20%, exactly how much additional profit would the business have kept? This gives leadership a concrete number to act on.

**D — Predictive modelling** is the scikit-learn phase. A Random Forest regressor takes all features simultaneously — Discount, Category, Region, Segment, Sub-Category, Ship Mode — and learns which combination of factors best predicts profit margin. The feature importance output will either confirm what EDA found (discounting dominates) or reveal something the correlation analysis missed.
