You can click any question card to jump straight into that analysis. Here's the reasoning behind the priority ordering:

The two profitability questions and the two discount questions are marked High priority because Phase 1 already told us something is structurally wrong — 18.7% loss-making rows and an 80% max discount are red flags that need explaining before anything else. These questions have the highest chance of producing an actionable recommendation.

The geography and customer segment questions are also High priority because they tell us *where* to focus any corrective action. Knowing that Tables are loss-making is useful; knowing that Tables sold in the Central region to Consumer customers at 40% discount are the specific problem is actionable.

The time-series and ship mode questions are Medium — they provide important context (is the business improving? are operations adding cost?) but they won't reshape strategy the way the first four topic areas will.

**Recommended EDA sequence:**

Starting with profitability by category → then discount impact → then geography → then customer segments → then time trends. This order builds from "what's broken" to "where is it broken" to "who is causing it" to "is it getting worse."
